```python
import requests
import pandas as pd

cities = {
    "Delhi": (28.6139, 77.2090),
    "Mumbai": (19.0760, 72.8777),
    "Kolkata": (22.5726, 88.3639),
    "Chennai": (13.0827, 80.2707),
    "Bangalore": (12.9716, 77.5946),
    "Hyderabad": (17.3850, 78.4867),
    "Bhilai": (21.2167, 81.4333)
}

rows = []

for city, (lat, lon) in cities.items():

    url = f"https://archive-api.open-meteo.com/v1/archive?latitude={lat}&longitude={lon}&start_date=2026-04-01&end_date=2026-04-20&hourly=temperature_2m,relativehumidity_2m,windspeed_10m,pressure_msl"
    
    data = requests.get(url).json()

    hourly = data["hourly"]

    for i in range(len(hourly["time"])):
        row = {
            "DateTime": hourly["time"][i],
            "City": city,
            "Temperature": hourly["temperature_2m"][i],
            "Humidity": hourly["relativehumidity_2m"][i],
            "Wind_Speed": hourly["windspeed_10m"][i],
            "Pressure": hourly["pressure_msl"][i]
        }
        rows.append(row)

weather_df = pd.DataFrame(rows)

weather_df.to_csv("weather_history.csv", index=False)

print("✅ Historical weather data saved")
```

```python
pollution_df["DateTime"] = pd.to_datetime(pollution_df["DateTime"])
weather_df["DateTime"] = pd.to_datetime(weather_df["DateTime"])
```

```python
pollution_df["DateTime"] = pollution_df["DateTime"].dt.floor("h")
weather_df["DateTime"] = weather_df["DateTime"].dt.floor("h")
```

```python
merged_df = pd.merge(pollution_df, weather_df, on=["City", "DateTime"], how="inner")
```

```python
import os

file_path = os.path.join(os.path.expanduser("~"), "Downloads", "final_dataset.csv")

merged_df.to_csv(file_path, index=False)

print("Saved at:", file_path)
```
